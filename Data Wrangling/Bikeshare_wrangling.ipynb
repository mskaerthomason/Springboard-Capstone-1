## Wrangling bikeshare data for Capstone 1

## Read in the data

### Let's start with reading in a test file of bikeshare data, Quarter 1 in 2016.

```python
import pandas as pd
dateparse = lambda x: pd.datetime.strptime(x, '%m/%d/%Y %H:%M')
bikeshare = pd.read_csv('../Data Wrangling/data/2016-Q1-Trips-History-Data.csv', parse_dates=True, date_parser=dateparse)
bikeshare
```

## Review the data

### Column names are easily readable, but not great for reference in code, so let's simplify.

```python
bikeshare.columns = ['duration', 'startdate','enddate','startlocID','startloc','endlocID','endloc','bikeID','memtype']
```

```python
bikeshare.info()
```

Those durations are in MILLISECONDS (really?!).  Let's convert to minutes.

```python
bikeshare.duration = bikeshare.duration * 1.66667e-5
```

```python
bikeshare['duration'].head()
```

### Are there missing values? No!

```python
bikeshare.isnull().values.any()
```

### Are there outliers? What is the distribution of ride duration?

```python
bikeshare.describe()
```

## Visualize the data

```python
import matplotlib.pyplot as plt
bikeshare.duration.plot('hist')
plt.show()
```

```python
bikeshare.plot(y='duration', kind='hist', bins=50, range=(0,200))
plt.show()
```

### Are there different distributions for registered vs. casual members?

```python
bikeshare.boxplot(column='duration', by='memtype')
plt.show()
```

### Let's parse by dates and times, too.

```python
bikeshare.plot(x='startdate',y='duration', kind='scatter')
plt.show()
```
